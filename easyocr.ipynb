```python
# Tokenization On image
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(image_text)
image_tokens = [w for w in word_tokens if not w.lower() in stop_words]
print(image_tokens)
```

```python
# Tokenization On pdf
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(pdf_text)
pdf_tokens = [w for w in word_tokens if not w.lower() in stop_words]
print(pdf_tokens)
```

```python
feature_names = []
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def isPresent(token):
    digits = "0123456789.-,'()!:;+â€˜%$#@&^*|"
    present = 0
    for i in range(len(token)):
        if token[i] in digits:
            present = 1
            return True
    return False
 

def final_tokens(token_list):
    res = []
    list = token_list
    for token in list:
        token = str(token)
        if not isPresent(token):
            if spell.correction(token) == token:
                if token.lower() not in feature_names:
                    feature_names.append(token.lower())

                if token.lower() not in res:
                    res.append(token.lower())

    final__tokens = res

    return final__tokens
```

```python
import os
target_labels = []
path_to_dataset = './data/'
```

```python
for entry in os.scandir(path_to_dataset):
    if entry.is_dir():
        target_labels.append(entry.path[7:])

target_labels
```

```python
def readPdf(file_path):
    input_pdf = open(file_path, 'rb')
    reader = PdfReader(input_pdf)
    page = reader.pages[0]
    pdf_text = page.extract_text()
    
    word_tokens = word_tokenize(pdf_text)
    pdf_tokens = [w for w in word_tokens if not w.lower() in stop_words]
    final_pdf_tokens = final_tokens(pdf_tokens)
    return final_pdf_tokens
```

```python
def readImage(image_path):
#     input_image = Image.open(image_path)
    images = [keras_ocr.tools.read(image_path)]
    prediction_groups = pipeline.recognize(images)
    predicted_image_1 = prediction_groups[0]
    image_text_list = []
    for text, box in predicted_image_1:
        image_text_list.append(text)
    image_text = " "
    image_text = image_text.join(image_text_list)
    
    word_tokens = word_tokenize(image_text)
    image_tokens = [w for w in word_tokens if not w.lower() in stop_words]
    final_image_tokens = final_tokens(image_tokens)
    return final_image_tokens
```

```python
for entry_in_dataset in os.scandir(path_to_dataset):
    for entry_in_file in os.scandir(entry_in_dataset):
        path_to_file = entry_in_file.path.replace('\\', '/')
        # now reading all dataset and adding the unique features into feature names
        if path_to_file[-4:] == '.pdf':
            #pdf input
            readPdf(path_to_file)
        elif path_to_file[-4:] == '.png' or path_to_file[-4:] == '.jpg':
            #image input
            readImage(path_to_file)
```

```python
import pandas as pd
df = pd.DataFrame(columns = feature_names)
```

```python
for entry_in_dataset in os.scandir(path_to_dataset):
    if entry_in_dataset.is_dir():
        className = entry_in_dataset.path[7:]
        print(className)
    for entry_in_file in os.scandir(entry_in_dataset):
        path_to_file = entry_in_file.path.replace('\\', '/')
        # now reading all dataset and adding the unique features into feature names
        if path_to_file[-4:] == '.pdf':
            #pdf input
            final_pdf_tokens = readPdf(path_to_file)
            row_for_df = [0] * len(feature_names)
            
            for i in range(len(feature_names)-1):
                if feature_names[i] in final_pdf_tokens:
                    row_for_df[i] = 1
            row_for_df[-1] = className
            df.loc[len(df.index)] = row_for_df
        elif path_to_file[-4:] == '.png' or path_to_file[-4:] == '.jpg':
            #image input
            final_image_tokens = readImage(path_to_file)
            row_for_df = [0] * len(feature_names)
            
            for i in range(len(feature_names)-1):
                if feature_names[i] in final_image_tokens:
                    row_for_df[i] = 1
            row_for_df[-1] = className
            df.loc[len(df.index)] = row_for_df
```
